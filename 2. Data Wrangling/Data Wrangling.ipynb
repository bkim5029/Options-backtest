```python
# only collects the columns we want to analyze and changes the format of the column to make it more readable
def call_df(dataframe):
    dataframe = dataframe[[' [QUOTE_DATE]', ' [EXPIRE_DATE]',
                           ' [UNDERLYING_LAST]', ' [DTE]',
                           ' [STRIKE]', ' [C_ASK]',
                           ' [C_BID]', ' [C_DELTA]',
                           ' [C_GAMMA]', ' [C_VEGA]',
                           ' [C_THETA]', ' [C_VOLUME]',
                           ' [C_IV]']]

    
    dataframe = dataframe.rename(columns = {' [QUOTE_DATE]':'QUOTE_DATE',                          
                                            ' [EXPIRE_DATE]':'EXPIRE_DATE', 
                                            ' [UNDERLYING_LAST]':'LAST_PRICE',
                                            ' [DTE]':'DTE',
                                            ' [STRIKE]':'STRIKE',
                                            ' [C_ASK]':'C_ASK',
                                            ' [C_BID]':'C_BID',
                                            ' [C_DELTA]':'C_DELTA',
                                            ' [C_GAMMA]':'C_GAMMA',
                                            ' [C_VEGA]':'C_VEGA',
                                            ' [C_THETA]':'C_THETA',
                                            ' [C_VOLUME]':'C_VOLUME',
                                            ' [C_IV]':'C_IV'
                                            })
    return dataframe
```

```python
def put_df(dataframe):
    dataframe = dataframe[[' [QUOTE_DATE]', ' [EXPIRE_DATE]',
                           ' [UNDERLYING_LAST]', ' [DTE]',
                           ' [STRIKE]', ' [P_ASK]',
                           ' [P_BID]', ' [P_DELTA]',
                           ' [P_GAMMA]', ' [P_VEGA]',
                           ' [P_THETA]', ' [P_VOLUME]',
                           ' [P_IV]']]

    dataframe = dataframe.rename(columns={' [QUOTE_DATE]': 'QUOTE_DATE',
                                          ' [EXPIRE_DATE]': 'EXPIRE_DATE',
                                          ' [UNDERLYING_LAST]': 'LAST_PRICE',
                                          ' [DTE]': 'DTE',
                                          ' [STRIKE]': 'STRIKE',
                                          ' [P_ASK]': 'P_ASK',
                                          ' [P_BID]': 'P_BID',
                                          ' [P_DELTA]': 'P_DELTA',
                                          ' [P_GAMMA]': 'P_GAMMA',
                                          ' [P_VEGA]': 'P_VEGA',
                                          ' [P_THETA]': 'P_THETA',
                                          ' [P_VOLUME]': 'P_VOLUME',
                                          ' [P_IV]': 'P_IV'
                                          })
    return dataframe
```

```python
aapl_2017_df_put = put_df(aapl_2017_df)
aapl_2018_df_put = put_df(aapl_2018_df)
aapl_2019_df_put = put_df(aapl_2019_df)
aapl_2020_df_put = put_df(aapl_2020_df)
aapl_put_df = put_df(aapl_df)
```

Creating a df that meets my criteria of trading option contract

```python
trading_df = pd.DataFrame()
for rows in range(len(aapl_put_df)):
    if (30 <= aapl_put_df['PROB_ITM'][rows] <= 34) and (38 <= aapl_put_df['DTE'][rows] <= 45):
        trade_row = (aapl_put_df.iloc[[rows]])
        trading_df = pd.concat([trading_df,trade_row])
trading_df = trading_df.reset_index(drop = True)
trading_df
```

Any rows with same `EXPIRE_DATE` is considered same option trade because they are same contract just read on different date.

We only want 1 contract per day so we will
- only retrieve one unique expire date
- only retrieve one unique strike

ex. index 5,6,7 are the same trade but read on different date

### Deleting the dupe EXPIRE_DATE

```python
seen_expire = []
my_trade = pd.DataFrame(columns = ['QUOTE_DATE','EXPIRE_DATE','LAST_PRICE','DTE','STRIKE','P_ASK','P_BID','P_DELTA','P_GAMMA','P_VEGA','P_THETA','P_VOLUME','P_IV'])
for i in range(len(trading_df)):
    expire_date = trading_df['EXPIRE_DATE'][i]
    if expire_date not in seen_expire:   
        seen_expire.append(expire_date)
        trade_row = trading_df.iloc[[i]]
        my_trade = pd.concat([my_trade, trade_row])
        my_trade = my_trade.reset_index(drop=True)
    else:
        pass
my_trade
```

each row of `my_df` is our option trade and each row is needed to find the full contract of the data

in order to harvest full historic option contract data, we need
- EXPIRE_DATE
- STRIKE
- DTE

```python
latest_expire = my_trade.iloc[[0]]['EXPIRE_DATE'].tolist()[0]
latest_strike = my_trade.iloc[[0]]['STRIKE'].tolist()[0]
latest_dte = my_trade.iloc[[0]]['DTE'].tolist()[0]
aapl_put_df.loc[(aapl_put_df['EXPIRE_DATE'] == latest_expire) & (aapl_put_df['STRIKE'] == latest_strike) & (aapl_put_df['DTE'] <= latest_dte)]
```

Because our data is limited, some contracts will not have full data so we will only collect the contract that has full data

```python
complete_c = []
not_complete_c = []
for i in range(len(my_trade)):
    expire = my_trade.iloc[[i]]['EXPIRE_DATE'].tolist()[0]
    strike = my_trade.iloc[[i]]['STRIKE'].tolist()[0]
    dte = my_trade.iloc[[i]]['DTE'].tolist()[0]
    
    f_trade_df = aapl_put_df.loc[(aapl_put_df['EXPIRE_DATE'] == expire) & (aapl_put_df['STRIKE'] == strike) & (aapl_put_df['DTE'] <= dte)]
    f_trade_df = f_trade_df.reset_index(drop=True)
    if (f_trade_df['DTE'].tail(1).tolist()[0] == 0) or (f_trade_df['DTE'].tail(1).tolist()[0] == 1) or (f_trade_df['DTE'].tail(1).tolist()[0] == 2): # I only want df that contains full data
        complete_c.append(f_trade_df)
    else:
        not_complete_c.append(f_trade_df)
```
