Takes string date and adds number of days that is given and outputs datetime object

```python
def add_days_to_date(original_date, days_added):
    output = datetime.strptime(original_date, '%Y-%m-%d')
    output = output + timedelta(days = days_added)            
    date_string = datetime.strftime(output, '%Y-%m-%d')   
    return output
```

returns a list of historic option df that only has complete data

```python
complete_c = []
for i in range(len(my_trade)):
    expire = my_trade.iloc[[i]]['EXPIRE_DATE'].tolist()[0]
    strike = my_trade.iloc[[i]]['STRIKE'].tolist()[0]
    dte = my_trade.iloc[[i]]['DTE'].tolist()[0]
    
    f_trade_df = aapl_put_df.loc[(aapl_put_df['EXPIRE_DATE'] == expire) & (aapl_put_df['STRIKE'] == strike) & (aapl_put_df['DTE'] <= dte)]
    f_trade_df = f_trade_df.reset_index(drop=True)
    if (f_trade_df['DTE'].tail(1).tolist()[0] == 0) or (f_trade_df['DTE'].tail(1).tolist()[0] == 1) or (f_trade_df['DTE'].tail(1).tolist()[0] == 2): # I only want df that contains full data
        complete_c.append(f_trade_df)
```

gives list of different strike df of our trade option

```python
# this df allows us to find the ideal verticle strike
my_verticle = []
for i in range(len(my_trade)):
    latest_expire = my_trade.iloc[[i]]['EXPIRE_DATE'].tolist()[0]
    latest_quote_date = my_trade.iloc[[i]]['QUOTE_DATE'].tolist()[0]
    latest_dte = my_trade.iloc[[i]]['DTE'].tolist()[0]
    # to find the data we need, all these three criteria has to be matched in our aapl_put_df
    f_trade_df = aapl_put_df.loc[(aapl_put_df['EXPIRE_DATE'] == latest_expire) & (aapl_put_df['QUOTE_DATE'] == latest_quote_date) & (aapl_put_df['DTE'] <= latest_dte)]
    # only collect data that has PROB_ITM less than my_trade's PROB_ITM 
    f_trade_df = f_trade_df.loc[f_trade_df['PROB_ITM'] <= my_trade['PROB_ITM'][i]].sort_values(by='PROB_ITM', ascending=False)
    f_trade_df = f_trade_df.reset_index(drop=True)
    my_verticle.append(f_trade_df)
```
