```python
study_df = aapl_put_df.loc[(aapl_put_df['EXPIRE_DATE'] == '2017-02-10') & (aapl_put_df['STRIKE'] == 112) & (aapl_put_df['DTE'] <= 38)]
study_df = study_df.reset_index(drop=True)
study_df
```

profit df analysis

```python
def profit_df(df):
    p_mid_list = []
    for row in range(len(df)):
        p_mid = ((df['P_ASK'][row] + df['P_BID'][row]) / 2).round(2)
        p_mid_list.append(p_mid)

    amount_rec = p_mid_list[0]
    profit_list = []
    for pmid in p_mid_list:
        profit = (amount_rec - pmid).round(2)
        profit_list.append(profit)

    profit_perc_list = []
    for profit in profit_list:
        profit_perc = ((profit / p_mid_list[0]).round(2)) * 100
        profit_perc_list.append(profit_perc)
        
    output = df[['QUOTE_DATE', 'LAST_PRICE', 'DTE']]
    output = output.assign(P_MID = p_mid_list, PROFIT = profit_list, PROFIT_PERC = profit_perc_list)
    return output
```

```python
profit_df(study_df)
```

In this situation, we could have gotten out of this trade and moved on at `2017-01-09` which is only 6 days after our entry date.

Because this trade will held till DTE, it took more a month for it to gain another 50% of the profit.

Even worse, at `2017-02-01` our profit was at 99% and we waited 9 days to get the last 1% of our trade.

- If we divide the length of our trade and if the profit target of 50% has been reached before the half mark, we should move on with another trade because
    - It is not worth waiting time to get another 50% of our profit
    - Possibility of turning a winning trade into a losing trade. Lock in the winning trade and move on
 
# How many trades did we hit the 50% profit mark before the half mark?

```python
profit_put_sell_study_dict['naked_dte']
```

```python
before_50_date = []
for contract in range(len(profit_put_sell_study_dict['naked_dte'])):
    my_trade_expire = profit_put_sell_study_dict['naked_dte']['EXPIRE_DATE'][contract]
    my_trade_strike = profit_put_sell_study_dict['naked_dte']['STRIKE'][contract]
    my_trade_dte = profit_put_sell_study_dict['naked_dte']['DTE'][contract]
    
    each_trade = aapl_put_df.loc[(aapl_put_df['EXPIRE_DATE'] == my_trade_expire) & (aapl_put_df['STRIKE'] == my_trade_strike) & (aapl_put_df['DTE'] <= my_trade_dte)]
    each_trade = each_trade.reset_index(drop=True)
    transformed_df = profit_df(each_trade)
    half_mark = round(len(each_trade) / 2)
    
    for row in range(len(transformed_df)):
        if (row <= half_mark and transformed_df['PROFIT_PERC'][row] >= 50):
            before_50_date.append(transformed_df['QUOTE_DATE'][0])
            break
            
after_50_df = pd.DataFrame()
before_50_df = pd.DataFrame()
for date in profit_put_sell_study_dict['naked_dte']['TRADE_DATE']:
    if date in before_50_date:
        before_50_df = pd.concat([before_50_df, profit_put_sell_study_dict['naked_dte'].loc[profit_put_sell_study_dict['naked_dte']['TRADE_DATE'] == date]])
        before_50_df = before_50_df.reset_index(drop=True)
    else:
        after_50_df = pd.concat([after_50_df, profit_put_sell_study_dict['naked_dte'].loc[profit_put_sell_study_dict['naked_dte']['TRADE_DATE'] == date]])
        after_50_df = after_50_df.reset_index(drop=True)
```

# TAKEAWAY
Exit at 50% profit
- Gives us a chance to lock in our profit and move onto the next trade
- As soon as it hit our profit mark, we can exit the trade letting us trade other options
- Most of the trades hit the 50% mark before the half mark so its a wise choice to leave early rather than waiting time to recieve the other 50%

Exit at DTE
- Should `CONSIDER` using when our 50% profit mark hasnt been reached before the half mark
    - After the halfmark, it is traders option to stay in or exit the trade
- Not efficent to wait long period of time to get the full profit
