```python
import pandas as pd
```

```python
naked_put_sell_50_study_df = pd.read_csv('C:\\Users\\bkim5\\EVERYTHING CODING\\PROJECT\\Options project\\3. Option Studies\\naked_put_sell_50_study.csv')
naked_put_sell_dte_study_df = pd.read_csv('C:\\Users\\bkim5\\EVERYTHING CODING\\PROJECT\\Options project\\3. Option Studies\\naked_put_sell_dte_study.csv')
verticle_put_sell_50_study_df = pd.read_csv('C:\\Users\\bkim5\\EVERYTHING CODING\\PROJECT\\Options project\\3. Option Studies\\verticle_put_sell_50_study.csv')
verticle_put_sell_dte_study_df = pd.read_csv('C:\\Users\\bkim5\\EVERYTHING CODING\\PROJECT\\Options project\\3. Option Studies\\verticle_put_sell_dte_study.csv')
put_sell_study_dict = {'naked_50': naked_put_sell_50_study_df,
                       'naked_dte' : naked_put_sell_dte_study_df,
                       'verticle_50' : verticle_put_sell_50_study_df,
                       'verticle_dte' : verticle_put_sell_dte_study_df}
```

```python
def profit(df):
    output = df.copy()
    for row in range(len(output)):
        if output['PROFIT'][row] < 0:
            output = output.drop(row)
    output = output.reset_index(drop=True)
    return output
```

```python
profit(naked_put_sell_dte_study_df)
```

```python
profit_naked_put_sell_50_study_df = profit(naked_put_sell_50_study_df)
profit_naked_put_sell_dte_study_df = profit(naked_put_sell_dte_study_df)
profit_verticle_put_sell_50_study_df = profit(verticle_put_sell_50_study_df)
profit_verticle_put_sell_dte_study_df = profit(verticle_put_sell_dte_study_df)
profit_put_sell_study_dict = {'naked_50': profit_naked_put_sell_50_study_df,
                              'naked_dte' : profit_naked_put_sell_dte_study_df,
                              'verticle_50' : profit_verticle_put_sell_50_study_df,
                              'verticle_dte' : profit_verticle_put_sell_dte_study_df}
```

```python
profit_put_sell_study_dict['naked_dte']
```

```python
def loss(df):
    output = df.copy()
    for row in range(len(output)):
        if output['PROFIT'][row] > 0:
            output.drop(row, inplace=True)
    output = output.reset_index(drop=True)
    return output
```

```python
loss_naked_put_sell_50_study_df = loss(naked_put_sell_50_study_df)
loss_naked_put_sell_dte_study_df = loss(naked_put_sell_dte_study_df)
loss_verticle_put_sell_50_study_df = loss(verticle_put_sell_50_study_df)
loss_verticle_put_sell_dte_study_df = loss(verticle_put_sell_dte_study_df)
loss_put_sell_study_dict = {'naked_50': loss_naked_put_sell_50_study_df,
                              'naked_dte' : loss_naked_put_sell_dte_study_df,
                              'verticle_50' : loss_verticle_put_sell_50_study_df,
                              'verticle_dte' : loss_verticle_put_sell_dte_study_df}
```
