# Why did naked_50 had the highest profit overall?
- naked_50 is a high risk high reward play because you are not sacrificing your profit to buy insurance.
- because naked trades have unlimited risk, when the stock decrease, your naked option will also decrease unlimitly
- In this situation, for the past 3 years, the market was bullish and helped naked position.
- Lets simulate a bearish market by only collecting data when the stock price when down

```python
def bear_df(df):
    output_df = pd.DataFrame()
    for row in range(len(df)):
        if (df['TRADE_LAST_PRICE'][row] > df['EXIT_LAST_PRICE'][row]):
            output_df = pd.concat([output_df, df.iloc[[row]]])
            output_df = output_df.reset_index(drop=True)
    return output_df
```

```python
naked_50_bear = bear_df(put_sell_study_dict['naked_50'])
naked_50_bear
```

```python
vertical_50_bear = bear_df(put_sell_study_dict['verticle_50'])
vertical_50_bear
```

```python
naked_dte_bear = bear_df(put_sell_study_dict['naked_dte'])
naked_dte_bear
```

```python
vertical_dte_bear = bear_df(put_sell_study_dict['verticle_dte'])
vertical_dte_bear
```

In a bearish market, naked trades have extreme losses while vertical trades have mitigated the loss due to its insurance

# Takeaway
Vertical
- Vertical trades are considerd `safe` because you are protected against extreme down movement.
    - You are sacrificing some of your profit to buy this insurance
    
    - Vertical should be used if you are unsure of the stock direction and want a protection
    
Naked
- Naked trades are considered `dangerous` because you are unprotected against extreme down movement.
    - You are not buying any insurance for your option position so you are getting the full profit
    
    - Naked should be used if you are confident that stock direction will go your way.
    
There is no way people can predict the direction of a market. For that reason, Vertical trades are optimal for most traders and Naked for people who are able to take lots of risk.
